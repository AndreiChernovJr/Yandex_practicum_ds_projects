# Выбор локации для скважины

Допустим, мы работаем в добывающей компании «ГлавРосГосНефть». Нужно решить, где бурить новую скважину.

Нам предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов. Постройте модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Нам предстоит проанализировать возможную прибыль и риски техникой *Bootstrap.*

Шаги для выбора локации:

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

Для каждого региона все будем считать отдельно.

```python
BUDGET = 10000000
UNIT_PRICE = 450
MINES = 200
NO_LOSS = (BUDGET/200)/UNIT_PRICE
print('объем нефти в средней скважине для безубыточности', NO_LOSS)
```

Тем не менее вторая модель самая лучшая, а это значит, что нас ждет увлекательное противостояние более богатых регионов и более качественной модели. Будет интересно понаблюдать, как эти факторы отразятся на потенциальной прибыли регионов.

```python
def economy(model, features, target, budget, unit_price, mines):
    predictions = model.predict(features)
    predictions = pd.Series(predictions)
    predictions_sorted = predictions.sort_values(ascending=False)
    target = pd.Series(target).reset_index(drop=True)
    selected = target[predictions_sorted.index][:mines]
    profit = selected.sum() * unit_price - budget
    return profit
```

Для начала мы сделали масштабирование, но оно не дало никакого эффекта, поэтому мы вернем расчеты к оригинальным. И приступим к задумке №2 - сделать отдельную модель для каждого региона.

```python
bootstrap_samples = 1000
profit0 = []
loss_count = 0
for i in range(bootstrap_samples):
    subsample = features_valid0.sample(n=500, replace=True, random_state=i)
    target_subs = target_valid0.sample(n=500, replace=True, random_state=i)
    result = economy(model0, subsample, target_subs, BUDGET, UNIT_PRICE, MINES)
    profit0.append(result)
    if result <=0:
        loss_count += 1
profit0 = pd.Series(profit0)
print(loss_count/bootstrap_samples, profit0.mean())
if loss_count>=25:
    print('добыча в регионе рискована')
```

confidence_interval0 = stats.t.interval(

```python
lower0 = profit0.quantile(0.025)
upper0 = profit0.quantile(0.975)

print(lower0, upper0)
```
