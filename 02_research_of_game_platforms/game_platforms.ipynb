<a id='pt2_5'></a>

```python
multiplatform = data.groupby('name')['platform'].count()

def multicheck(name):
    if multiplatform[name]>1:
        return True
    if multiplatform[name]==1:
        return False        
data['multiplatform'] = data['name'].apply(multicheck)
print(data['multiplatform'].value_counts())
```

# Проверка Гипотез
<a id='pt5'></a>

```python
data_1.info()
```

Для начала избавимся от нулей в оценках, потому что проверка гипотезы по таким данным недопустима.

```python
#data_2 = data_1.drop(data_1[data_1['user_score']==0].index)
#print(data_2.shape)
```

```python
data_2 = data_1.dropna(subset=['user_score'])
print(data_2.shape)
```

1. H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
2. H1: Средние пользовательские рейтинги платформ Xbox One и PC не одинаковые
<a id='pt5_1'></a>

```python
data_2_xbox1 = data_2.loc[data_2['platform']=='XOne']
xbox1 = data_2_xbox1['user_score']
xbox1 = xbox1.reset_index(drop=True)
print(xbox1)
```

```python
data_2_pc = data_2.loc[data_2['platform']=='PC']
pc = data_2_pc['user_score']
pc = pc.reset_index(drop=True)
print(pc)
```

```python
alpha = .05
results = st.ttest_ind(xbox1, pc, equal_var = False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

А с новыми данными -новый результат. Наш p-value выше заданного уровня статистической значимости, а значит, что средние пользовательские рейтинги платформ Xbox One и PC могут быть одинаковыми.

<div class="alert alert-success">
<b>ОТЛИЧНО! 👍</b>

Верно то, что мы используем здесь двусторонний тест.

Ну и отлично, что ты понимаешь, что нельзя сделать вывод, что тест "подтвердил" одну из гипотез (в статистике гипотезы подтвердить невозможно) или "подтвердил" равенство/неравенство (можно говорить только о статистически (не)значимом различии). Например так: «У нас нет оснований утверждать, что средние пользовательские рейтинги платформ Xbox One и PC значимо различаются».</div>


1. Н0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.
2. Н1: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
<a id='pt5_2'></a>

```python
data_2_action = data_2.loc[data_2['genre']=='Action']
action = data_2_action['user_score']
action = action.reset_index(drop=True)
print(action)
```

```python
data_2_sports = data_2.loc[data_2['genre']=='Sports']
sports = data_2_sports['user_score']
sports = sports.reset_index(drop=True)
print(sports)
```

```python
alpha = .05
results = st.ttest_ind(action, sports, equal_var = False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

А здесь мы сожем смело отвергнуть гипотезу, что средние пользовательские рейтинги жанров Action и Sports могут быть одинаковыми. Не могут! Шансы просто микроскоопические.
